```python
import os
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile

metaDir = os.path.join(os.path.dirname('__file__'), '..', 'metadata')
p4Dir = os.path.join(os.path.dirname('__file__'), '..', '4')
 
df = pd.read_excel(os.path.join(metaDir, 'Gene_Map.xlsx'), sheetname='Map')
  
    
# Getting the Gene Map
gene_list = list(df['Gene'])
uniprot_list = list(df['UniprotAC'])
gene_id_list = list(df['GeneID'])
```

```python
df1 = pd.read_csv(os.path.join(p4Dir, '4.1.csv'))
df2 = pd.read_csv(os.path.join(p4Dir, '4.2.csv'))
df3 = pd.read_csv(os.path.join(p4Dir, '4.3.csv'))

bioList1 = []
bioList1 = df1["GeneA"]
bioList1 = set(bioList1.append(df1["GeneB"]))

if 'Missing' in bioList1: bioList1.remove('Missing')


gene_id_list_1 = []
for g in bioList1:
    index = gene_list.index(g)
    gene_id_list_1.append(gene_id_list[index])

#print(gene_id_list_1)

id_41 = pd.DataFrame(gene_id_list_1)

id_41.to_csv('4.1_GO.csv',encoding='utf8', index=False, header=False)


bioList2 = []
bioList2 = df2["GeneA"]
bioList2 = set(bioList2.append(df2["GeneB"]))

if 'Missing' in bioList2: bioList2.remove('Missing')

gene_id_list_2 = []
for g in bioList2:
    index = gene_list.index(g)
    gene_id_list_2.append(gene_id_list[index])

#print(gene_id_list_2)
    
id_42 = pd.DataFrame(gene_id_list_2)

id_42.to_csv('4.2_GO.csv',encoding='utf8', index=False, header=False)


bioList3 = []
bioList3 = df1["GeneA"]
bioList3 = set(bioList3.append(df3["GeneB"]))

if 'Missing' in bioList3: bioList3.remove('Missing')

gene_id_list_3 = []
for g in bioList3:
    index = gene_list.index(g)
    gene_id_list_3.append(gene_id_list[index])

#print(gene_id_list_3)
    
id_43 = pd.DataFrame(gene_id_list_3)

id_43.to_csv('4.3_GO.csv',encoding='utf8', index=False, header=False)
```
